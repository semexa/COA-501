```python
import cv2

imagen = cv2.imread('2022_0810_073735_037.jpg')

imagen2 = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY )
#cv2.imshow('prueba de imagen', imagen2)
cv2.imwrite('NUEVA2_GRAY.JPG', imagen2)

imagen3 = cv2.cvtColor(imagen, cv2.COLOR_BGR2BGRA )
#cv2.imshow('prueba de imagen', imagen2)
cv2.imwrite('NUEVA2_BGRA.JPG', imagen3)

imagen4 = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB )
#cv2.imshow('prueba de imagen', imagen2)
cv2.imwrite('NUEVA2_RGB.JPG', imagen4)

imagen5 = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGBA )
#cv2.imshow('prueba de imagen', imagen2)
cv2.imwrite('NUEVA2_RGBA.JPG', imagen5)

imagen6 = cv2.cvtColor(imagen, cv2.COLOR_BGR2YCrCb )
#cv2.imshow('prueba de imagen', imagen2)
cv2.imwrite('NUEVA2_YCrCb.JPG', imagen6)

imagen7 = cv2.cvtColor(imagen, cv2.COLOR_BGR2HSV )
#cv2.imshow('prueba de imagen', imagen2)
cv2.imwrite('NUEVA2_HSV.JPG', imagen7)

imagen8 = cv2.cvtColor(imagen, cv2.COLOR_BGR2HSV_FULL )
#cv2.imshow('prueba de imagen', imagen2)
cv2.imwrite('NUEVA2_HSV_FULL.JPG', imagen8)

imagen9 = cv2.cvtColor(imagen, cv2.COLOR_BGR2Lab )
#cv2.imshow('prueba de imagen', imagen2)
cv2.imwrite('NUEVA2_Lab.JPG', imagen9)

imagen10 = cv2.cvtColor(imagen, cv2.COLOR_BGR2Luv )
#cv2.imshow('prueba de imagen', imagen2)
cv2.imwrite('NUEVA2_Luv.JPG', imagen10)

imagen11 = cv2.cvtColor(imagen, cv2.COLOR_BGR2HLS )
#cv2.imshow('prueba de imagen', imagen2)
cv2.imwrite('NUEVA2_HLS.JPG', imagen11)

imagen12 = cv2.cvtColor(imagen, cv2.COLOR_BGR2HLS_FULL )
#cv2.imshow('prueba de imagen', imagen2)
cv2.imwrite('NUEVA2_HLS_FULL.JPG', imagen12)

imagen13 = cv2.cvtColor(imagen, cv2.COLOR_BGR2YUV )
#cv2.imshow('prueba de imagen', imagen2)
cv2.imwrite('NUEVA2_YUV.JPG', imagen13)

imagen14 = cv2.cvtColor(imagen, cv2.COLOR_BGR2XYZ )
#cv2.imshow('prueba de imagen', imagen2)
cv2.imwrite('NUEVA2_XYZ.JPG', imagen14)


cv2.waitKey(0)
cv2.destroyAllWindows()
```
